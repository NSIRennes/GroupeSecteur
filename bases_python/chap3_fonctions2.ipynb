## A Variables locales à une fonction.

On veut définir une fonction factorielle de n qui calcule le produit des n premiers entiers. On peut le faire avec une variable locale à la fonction et une boucle.

```python
def factorielle(n):
    res = 1    # Variable locale qui affichera le résultat
    for i in range(2, n+1):
        res = res * i
    return res        
```

```python
factorielle(4)
```

La variable locale res est définie à l'intérieur de la fonction, au moment de son initialisation. On peut observer que la variable locale res n'existe pas à l'extérieur de la fonction factorielle.

```python
res = 42
print(factorielle(4))
res
```

La variable res de valeur 42 est une __variable globale__ .    

La variable res, __locale à la fonction__, n'existe que dans la fonction. 

```python
# Résolution du problème....
y = factorielle(4)
y
```

## B Sorties anticipées  
Voici une fonction qui teste si un nombre est premier ou non . 

```python
def est_premier(n):
    if n <= 1:                  # Cas particulier
        return False
    else:                       # Résolution du problème
        for i in range(2, n):
            if n%i == 0 : return False  # Le nombre n'est pas premier.
                                        # Le nombre est divisible par un entier 
                                        # compris entre 2 et n (non compris)
        return True                     # le nombre est premier
```

```python
est_premier(4)
```

```python
est_premier(7)
```

Le programme comprend :   
* Une instruction return dans une boucle, en interrompant l'exécution de la fonction, impliquant l'abandon des instructions restantes à l'intérieur du tour de boucle en cours ainsi que l'ensemble des tours de boucles restant à effectuer. Cela permet d'éviter d'effectuer les tours restants si le résultat est déja connu.
* Une instruction return en début de fonction qui permet d'écarter un cas particulier pour lequel le résultat est connu d'avance et qui permet de se concentrer ensuite sur le cas général du problème à résoudre.

__Les Docstrings__  

Python permet d’associer une documentation à toute fonction, sous la forme d’une chaîne de caractères donnée entre triple guillemets.

```python
def est_premier(n):
    """ Fonction qui prend en paramètre un nombre entier
        et renvoie  un booléen"""
    if n <= 1:                  # Cas particulier
        return False
    else:                       # Résolution du problème
        for i in range(2, n):
            if n%i == 0 : return False  # Le nombre n'est pas premier.
                                        # Le nombre est divisible par un entier 
                                        # compris entre 2 et n (non compris)
        return True                     # le nombre est premier
```

Cette chaîne est placée en début de fonction et peut contenir plusieurs lignes.  

On peut avoir accès à cette documentation avec la commande help de Python.

```python
help(est_premier)
```
